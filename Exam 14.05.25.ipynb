## Question 2 - Data Analysis

```python
#1. What is the average price of an item?
grouped = items.groupby('item_name')['item_price'].mean()
print(grouped)
```

```python
#2. Which customer has purchased the most products?
grouped1 = order_item.groupby('order_id')[['quantity']].sum()
customers = customers.rename(columns={'id':'customer_id'})
merged1 = pd.merge(grouped1,orders[['customer_id','order_id']], on = 'order_id', how = 'left')
merged1 = pd.merge(merged1,customers[['first_name','last_name', 'customer_id']], on = 'customer_id', how = 'left')
merged1[merged1['quantity'] == merged1['quantity'].max()]
```

```python
#3. Add a total_price column to the order_item table, calculating the purchase cost
items = items.rename(columns={'id':'item_id'})
order_item = pd.merge(order_item, items[['item_id', 'item_price']], on='item_id', how='left')
order_item['total_price'] = order_item['quantity'] * order_item['item_price']
display(order_item)
```

```python
#4. Display each purchase and its total price
grouped2 = order_item.groupby('order_id')[['total_price']].sum()
display(grouped2)
```

```python
#5. Find what is the most expensive purchase?
print("The most expensive:")
grouped2[grouped2['total_price'] == grouped2['total_price'].max()]
```

```python
#5. Find what is The cheapest purchase? ?
print("the chepest:")
filtered = grouped2[grouped2['total_price'] > 0]
filtered[filtered['total_price'] == filtered['total_price'].min()]
```

```python
#5. Find what is The average purchase?
mean_price = grouped2['total_price'].mean()
closest = grouped2.loc[(grouped2[['total_price']] - mean_price).abs().idxmin()]
print("the average:")
print(closest)
```

## Question 2 - Research And Present In A Graph

```python
customers1 = customers.dropna()
order_item1 = order_item.dropna()
orders1 = orders.dropna()
items1 = items.dropna()
```

```python
#1. Distribution of customers by gender
gender_count = customers1['gender'].value_counts()
gender_count.plot(kind = 'bar')
plt.title('Distribution of customers by gender')
plt.xlabel ('gender')
plt.xticks(rotation = 45)
plt.ylabel('customers count')
plt.tight_layout()
plt.show()
```

```python
#2. Distribution of purchase quantities by nationality
orders = pd.merge(orders,customers[['nationallity','customer_id']], on = 'customer_id', how = 'left')
grouped3 = orders.groupby('nationallity').size().reset_index(name='count')
sns.scatterplot(data=grouped3, x='nationallity', y='count')
plt.title('Purchases quantity by nationality')
plt.xlabel('Nationallity')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#3. Histogram of customer distribution by age
plt.hist(customers1['age'],edgecolor = 'Black')
plt.title('customers ages')
plt.xlabel('age')
plt.ylabel('customers count')
plt.tight_layout()
plt.show()
```

```python
#4. Graph showing the quantity of new customers who joined, by year
customers['joining_date'] = pd.to_datetime(customers['joining_date'], errors = 'coerce')
customers['join_year'] = customers['joining_date'].dt.year
customer_count = customers.groupby('join_year').size().reset_index(name = 'new_customer')
plt.plot(customer_count['join_year'], customer_count['new_customer'])
plt.title('New customers by year')
plt.xlabel('Year')
plt.ylabel('customer count')
plt.tight_layout()
plt.show()
```

```python
#5. Graph showing the quantity of sales by months
orders['order_date'] = pd.to_datetime(orders['order_date'], errors = 'coerce')
orders['order_month'] = orders['order_date'].dt.month
sale_count = orders.groupby('order_month').size().reset_index(name = 'new_order')
plt.plot(sale_count['order_month'], sale_count['new_order'])
plt.title('Sales by month')
plt.xlabel('month')
plt.ylabel('Sale count')
plt.tight_layout()
plt.show()
```

```python
#6. Histogram showing the distribution of purchase quantities by age (each purchase
#   should be calculated in a single row. No need to count product quantities, etc.) - Is
#   there more/less purchasing in certain age groups?
orders = pd.merge(orders,customers[['age','customer_id']], on = 'customer_id', how = 'left')
sns.histplot(orders['age'])

plt.title('Purchase quantities by customer age')
plt.xlabel('Age')
plt.ylabel('Purchase quantities')
plt.tight_layout()
plt.show()
```
